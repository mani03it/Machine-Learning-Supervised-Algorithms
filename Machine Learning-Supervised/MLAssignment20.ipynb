```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import load_boston
import seaborn as sns
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline

boston = load_boston()

bos = pd.DataFrame(boston.data)
bos.head()
```

```python
boston.feature_names
```

```python
print(boston.DESCR)
```

```python
bos_df = pd.DataFrame(boston.data, columns=boston.feature_names)
bos_df.head()
```

```python
bos_df['House_Price'] = boston.target
bos_df.head()
```

```python
bos_df.describe()
```

```python
from IPython.core.interactiveshell import InteractiveShell 
InteractiveShell.ast_node_interactivity = "all"
```

```python
x = bos_df.corr()
x
```

```python
plt.subplots(figsize=(20,20))
sns.heatmap(x, cmap='RdYlGn', annot= True)
plt.show()
```

```python
x = bos_df.drop('House_Price', axis=1)
y = bos_df['House_Price']
x.head()
y.head()
```

#### Building the model

```python
#Splitting train and test dataset
train_x, test_x, train_y, test_y = train_test_split(x,y,test_size=0.25, random_state=1)
train_x.shape
test_x.shape
train_y.shape
test_y.shape
```

```python
from sklearn.linear_model import LinearRegression
#Creating instance of LinearRegression
lm = LinearRegression()
#training the model
lm.fit(train_x, train_y)
```

```python
#Predicting using test dataset
pred_test = lm.predict(test_x)
```

```python
print(lm.coef_)
```

```python
df_model = pd.DataFrame({'Features':x.columns,'Coeff':lm.coef_})
df_model = df_model.sort_values(by=['Coeff'])
df_model
```

```python
df_model.plot(x='Features', y='Coeff',kind='bar', figsize=(15,10))
plt.show()
```

```python
print("R-Square value for Test Data is "+str(np.round(lm.score(test_x,test_y)*100, 0)))
print("R-Square value for Train Data is "+str(np.round(lm.score(train_x,train_y)*100, 0)))
```

```python
from sklearn import metrics

print("Mean Square Error(MSE) for test data is "+str(np.round(metrics.mean_squared_error(test_y, pred_test),0)))
print("Mean Absolute Error(MAE) for test data is "+str(np.round(metrics.mean_absolute_error(test_y, pred_test),0)))
```

```python
final_df = pd.concat([test_x, test_y],1)
final_df['Predicted'] = np.round(pred_test, 1)
final_df['Prediction_Error'] = final_df['House_Price'] - final_df['Predicted']
final_df
```

#### Linear Regression using statmodels 

```python
import statsmodels.formula.api as smf
bos_df.head()
```

```python
model = smf.ols('House_Price ~ CRIM+ZN+INDUS+CHAS+NOX+RM+AGE+DIS+RAD+TAX+PTRATIO+B+LSTAT', data=bos_df).fit()
print(model.summary())
```

```python
#Removing insignificant variables like INDUS and AGE
model = smf.ols('House_Price ~ CRIM+ZN+CHAS+NOX+RM+DIS+RAD+TAX+PTRATIO+B+LSTAT', data=bos_df).fit()
print(model.summary())
```

```python
#printing heteroscedasticity robust standard errors
h_model = model.get_robustcov_results()
print(h_model.summary())
```
